#### 2 Implementar a função que calcula o threshold global (ver slides da aula).

```octave
function threshold = thresholdlobal(img)
    threshold = 127;
    thresholdOld = 0;

    while abs (threshold - thresholdOld) > 0.1
        grup1 = img(img <= threshold);
        grup2 = img(img > threshold);
        average1 = mean(grup1);
        average2 = mean(grup2);
        thresholdOld = threshold;
        threshold = (average1 + average2) / 2;
    endwhile

endfunction
```
