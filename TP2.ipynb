```octave
function nova_imagem = transladar(img, v, h)
    [linha, coluna, ~] = size(img);
    nova_imagem = zeros(linha, coluna);
    slinha = 1; elinha = linha; ilinha = 1;
    scoluna = 1; ecoluna = coluna; icoluna = 1;
    if v > 0
    scoluna = coluna; ecoluna = 1; icoluna = -1;
    endif
    if h < 0
    slinha = linha; elinha = 1; ilinha = -1;
    endif
    for i = slinha : ilinha : elinha
    for j = scoluna : icoluna : ecoluna
      if i-h >= 1 && i-h <= linha && j+v >= 1 && j+v <= coluna
          nova_imagem(i-h, j+v) = img(i, j);  
      endif
    endfor
    endfor
    nova_imagem = uint8(nova_imagem);
end
```

```octave
imshow(transladar(im, 50,20));
```

```octave
function nova_imagem = rotacionar(img, ang)
  nova_imagem = zeros(size(img));
  
  midx=ceil(((size(img,1)+1)/2));
  midy=ceil(((size(img,2)+1)/2));
 
  for i=1:size(nova_imagem, 1)
    for j=1:size(nova_imagem, 2)
      x = (i-midx)*cosd(ang)-(j-midy)*sind(ang);
      y = (i-midx)*sind(ang)+(j-midy)*cosd(ang);
      x = round(x)+midx;
      y = round(y)+midy;
      
      if (x >= 1 && y >= 1 && x <= size(nova_imagem, 1) && y <= size(nova_imagem, 2))
        nova_imagem(x, y) = img(i, j);
      endif
    endfor
  endfor
  nova_imagem = uint8(nova_imagem);
endfunction
```

```octave
imshow(rotacionar(imread("lenna.png"), 30));
```

```octave
function nimg = cisalhamento(img, y)
  [lin, col, ~] = size(img);
  nimg = zeros(lin, col);
  mat_cisalhamento = [1 y 0; 0 1 0; 0 0 1];
  for i = 1:lin
    for j = 1:col
        c = mat_cisalhamento * [i; j; 1]
        c_i = floor(c(1))
        c_j = floor(c(2));
      if c_i > 0 && c_j > 0 && c_i <= lin && c_j <= col
        nimg(c_i, c_j) = img(i,j);
      endif
    endfor
 endfor
  
  nimg = uint8(nimg);
endfunction;
```

```octave
img_cisa = cisalhamento(im, -5);
```

```octave
function nova_imagem = escalamento(img , x, y)
  [coluna, linha, ~] = size(img);
  ncoluna = coluna * x;
  nlinha = linha * y;
  
  linhat = ceil([1:(size(img,1)*y)]./(y));
  cInt = ceil([1:(size(img,2)*x)]./(x));
  d = double(img);
  
  aux = d(:,:,1);
  R = aux(linhat, :);
  R = R(:, cInt);
  
  aux = d(:, :, 2);
  G = aux(linhat, :);
  G = G(:, cInt); 
  aux = d(:, :, 3);
  
  B = aux(linhat, :);
  B = B(:, cInt); 
  
  nova_imagem = zeros([nlinha, ncoluna, 3]);
  nova_imagem(:, :, 1) = R;
  nova_imagem(:, :, 2) = G;
  nova_imagem(:, :, 3) = B;
  
  nova_imagem = uint8(nova_imagem);
 endfunction;
```

```octave
imshow(escalamento(im, 3, 2));
```
